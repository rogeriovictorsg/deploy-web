#Construindo uma distribuição de frequência e um histograma de frequência 
Nos exercícios 31 a 34, construa uma distribuição de frequência e um histograma de frequência para o conjunto de dados usando o número de classes indicado. Descreva quaisquer padrões. Utilize R.

```
#31 Vendas.
v = c(2114,2468,7119,1876,4105,3183,1932,1355,4278,1030,2000,1077,5835,1512,1697,2478,3981,1643,1858,1500,4608,1000)
hist(v, nclass = 6)

freq_v = table(v)
barplot(freq_v)

#Podemos perceber que não há normalidade nos dados e que não há repetição de valores de vendas.
```

```
#32 Ardência de pimentas
a = c(35,51,44,42,37,38,36,39,44,43,40,40,32,39,41,38,42,39,40,46,37,35,41,39)
hist(a, nclass = 5)

freq_a = table(a)
barplot(freq_a)

#Percebemos que há uma tendência a normalidade de dados e que existem muitos valores repetidos
```

```
#33. Tempos de reação
t = c(507,389,305,291,336,310,514,442,373,428,387,454,323,441,388,426,411,382,320,450,309,416,359,388,307,337,469,351,422,412)
hist(t, nclass = 8)

freq_t = table(t)
barplot(freq_t)

#Percebemos que não há normalidade nos dados e que somente um valor se repete
```
